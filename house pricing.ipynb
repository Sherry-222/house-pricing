```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# -----------------------------
# Step 1: Dataset
# -----------------------------
# House Size (X) and Price (y)
X = np.array([500, 1000, 1500, 2000, 2500]).reshape(-1, 1)
y = np.array([25, 50, 75, 100, 125])   # in lakhs

# -----------------------------
# Step 2: Create and Train Model
# -----------------------------
model = LinearRegression()
model.fit(X, y)

# Equation of line
m = model.coef_[0]     # slope
c = model.intercept_   # intercept
print(f"Equation: Price = {m:.2f} * Size + {c:.2f}")

# Predict price for 2200 sq.ft
predicted_price = model.predict([[2200]])
print(f"Predicted Price for 2200 sq.ft = {predicted_price[0]:.2f} lakhs")

# -----------------------------
# Step 3: Plot Regression Line
# -----------------------------
line_X = np.linspace(400, 2600, 100).reshape(-1,1)
line_y = model.predict(line_X)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.scatter(X, y, color="blue", label="Data Points")
plt.plot(line_X, line_y, color="red", label="Regression Line")
plt.xlabel("Size (sq.ft)")
plt.ylabel("Price (Lakhs ₹)")
plt.title("House Size vs Price")
plt.legend()
plt.grid(True)

# -----------------------------
# Step 4: Simulate Epoch vs Accuracy
# -----------------------------
epochs = list(range(1, 11))
accuracy = [0.55,0.65,0.75,0.82,0.87,0.91,0.94,0.96,0.97,0.97]

plt.subplot(1,2,2)
plt.plot(epochs, accuracy, marker='o', color="green")
plt.xlabel("Epochs")
plt.ylabel("Accuracy (R² Score)")
plt.title("Epoch vs Accuracy")
plt.grid(True)

plt.tight_layout()
plt.show()
```
